<h3>Train-Test Split Using Sklearn</h3>

```python
# import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
car_df = pd.read_csv('carprices.csv')
car_df.head(10)
```

<h5>Mileage Vs Sell Price</h5>

```python
plt.scatter(car_df['Mileage'], car_df['Sell Price($)'])
```

<h5>Age Vs Sell Price</h5>

```python
plt.bar(car_df['Age(yrs)'], car_df['Sell Price($)'])
```

```python
x = car_df[['Mileage', 'Age(yrs)']]
x
```

```python
y = car_df['Sell Price($)']
y
```

<h5>Using Sklearn's Train-Test Split</h5>

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# Training Set of X
train_set_size_X = len(x_train)
train_set_size_X

# Testing Set of X
test_set_size_X = len(x_test)
test_set_size_X

# Training Set of Y
train_set_size_Y = len(y_train)
train_set_size_Y

# Testing Set of Y
test_set_size_Y = len(y_test)
test_set_size_Y
```

Total size = 19

Train set size of x and y = 16

Test set size of x and y = 4




<h5>Using Linear Regression Model</h5>

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
```

```python
clf.fit(x_train, y_train)
```

```python
# clf1.predict(x_test)
clf.predict(x_test)
```

<h5>Evaluating Accuracy Of The Linear Regression Model</h5>

```python
y_test
```

```python
clf_accuracy = clf.score(x_test, y_test) * 100
clf_accuracy
# Accuracy - 84.70%
```

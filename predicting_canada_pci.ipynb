<h3>Exercise 1: Predicting Canada's Per Capita Income (Linear Regression Model)</h3>

```python
# Importing Required Libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Uploading csv file

df_pci = pd.read_csv('canada_per_capita_income.csv')
df_pci.head(5)
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df_pci, figsize=(5,5))
```

<h4>Using Linear Regression To Predict Per Capita Income For The Year 2020</h4>

```python
reg1 = LinearRegression()
reg1.fit(df_pci[['year']], df_pci.per_capita_income)
# reg1.fit(df_pci[['per_capita_income']], df_pci.year)
reg1.predict([[2032]])
```

<p>Result: Predicted PCI Value for the year 2020 is 41288.69409442</p>
Similarly for the years,
<p>Year 2022: $42945.62424486</p>
<p>Year 2025: $45431.01947053</p>
<p>Year 2028: $47916.4146962</p> 
<p>Year 2032: $51230.27499709</p>

<h4>Getting an insight by using visualizing</h4>

```python
plt.xlabel('Year')
plt.ylabel('Per Capita Income ($)')
plt.scatter(df_pci.year, df_pci.per_capita_income, color = 'green', marker = '+')
plt.plot(df_pci.year, reg1.predict(df_pci[['year']]), color = 'red')
```
